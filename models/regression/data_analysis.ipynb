```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data_filepath = '../../data/csv/plant_data.csv'
```

```python
plant_data = pd.read_csv(data_filepath)
plant_data.tail()
```

```python
corr_matrix = plant_data.corr()
corr_matrix
```

```python
sn.heatmap(corr_matrix, annot=True, fmt=".1f", linewidths=.5)
plt.show()
```
